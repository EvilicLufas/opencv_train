```python
img = misc.imread('../data/lena.jpg')
```

```python
plt.figure(figsize=(10, 10))
for img_pyr_index, img_pyr in enumerate(pyramid_gaussian(img)):
#     print(img_pyr.shape)
    pyr_height, pyr_width, pyr_channel = img_pyr.shape
    if pyr_height < 30 or pyr_width < 30:
        break
    plt.subplot(1, 6, img_pyr_index+1)
    plt.imshow(img_pyr)
    plt.axis('off')
```
