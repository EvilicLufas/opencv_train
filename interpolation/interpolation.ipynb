```python
img = misc.imread('../data/blox.jpg')
# img = misc.imresize(img, (50, 50)) # width和height
```

```python
height, width, channel = img.shape
```

```python
size_lists = []
for size_range in range(1, 5):
    new_width = width*size_range
    new_height = height*size_range
    size_lists.append((new_width, new_height))
```

```python
interp_lists = ['nearest', 'lanczos', 'bilinear', 'bicubic', 'cubic']
```

```python
size_len = len(size_lists)
interp_len = len(interp_lists)
plt.figure(figsize=(10, 10))
for interp_index, interp_val in enumerate(interp_lists):
    for size_index, size_val in enumerate(size_lists):
        newimg = misc.imresize(img, size=size_val, interp=interp_val)
        plt.subplot(interp_len, size_len, size_len*interp_index + size_index+1)
        plt.imshow(newimg)
        plt.axis('off')
#         plt.title(str(size_val))
```
