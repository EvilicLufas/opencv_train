```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# cv2.ellipse?
cv2.putText?
```

```python
width = 512
height = 512
channel = 3
```

```python
img = np.zeros((width, height, channel), np.uint8)
```

```python
# plt.imshow(img)
```

```python
start_point = (0, 0)
end_point = (20, width)
line_color = (255, 0, 0) # RGB颜色
line_thick = 5
cv2.line(img, start_point, end_point, line_color, line_thick)
plt.imshow(img)
```

```python
left_upper_point = (30, 30)
right_down_point = (60, width)
line_color = (0, 255, 0) # RGB颜色
line_thick = 5
cv2.rectangle(img, left_upper_point, right_down_point, line_color, line_thick)
plt.imshow(img)
```

```python
center_point = (width/2, height/2)
radius = 20
line_color = (0, 0, 255) # RGB颜色
line_thick = 5
cv2.circle(img, center_point, radius, line_color, line_thick)
plt.imshow(img)
```

```python
axes_long = 100
axes_short = 50
line_color = (255, 0, 0) # RGB颜色
cv2.ellipse(img, (width/2,height/2), (axes_long, axes_short), 0, 0, 180, line_color, -1)
plt.imshow(img)
```

```python
center_point = (0+10, height/2-10)
radius = 20
line_color = (0, 0, 255) # RGB颜色
line_thick = 2
text_scale = 4
cv2.putText(img, 'OpenCV', center_point, cv2.FONT_HERSHEY_SIMPLEX, text_scale, line_color, line_thick)
plt.imshow(img)
```
