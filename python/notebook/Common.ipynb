# 计算梯度 $g_x$ 和 $g_y$ 的幅值和方向

```python
img = cv2.imread('../../data/lena.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = img.astype(np.float32)/255.0
img.dtype
```

```python
gx = cv2.Sobel(img, -1, dx=1, dy=0, ksize=1)
gy = cv2.Sobel(img, -1, dx=0, dy=1, ksize=1)
```

```python
magnitude, angle = cv2.cartToPolar(gx, gy, angleInDegrees=True)
```

```python
angle.max() # 在一般应用中使用无符号梯度方向，即180-x=x
```

```python
angle.min()
```

```python
plt.subplot(121)
plt.imshow(gx, cmap='gray') # gx由于是对水平相减，所以垂直方向敏感
plt.subplot(122)
plt.imshow(gy, cmap='gray') # gy由于是对垂直相减，所以水平方向敏感
```

```python
plt.subplot(121)
plt.imshow(magnitude, cmap='gray')
plt.subplot(122)
plt.imshow(angle, cmap='gray')
```
