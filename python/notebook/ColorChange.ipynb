```python
def rgb2gray(rgb):
    height, width, channel = rgb.shape
    assert channel==3
    gray = np.zeros((height, width), dtype=rgb.dtype)
    for row in range(height):
        for col in range(width):
            r = rgb[row, col, 0]
            g = rgb[row, col, 1]
            b = rgb[row, col, 2]
            gray[row, col] = int(0.299*r + 0.587*g + 0.114*b) # rgb转换为gray公式
    return gray
```
