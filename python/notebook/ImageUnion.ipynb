```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('../../data/lena.jpg')
```

```python
img_b = img[:, :, 0]
img_g = img[:, :, 1]
img_r = img[:, :, 2]
```

```python
plt.imshow(img_b)
```

```python
plt.imshow(img_g)
```

```python
plt.imshow(img_r)
```

```python
img_b, img_g, img_r = cv2.split(img) # 使用cv2分割img为BGR通道
```

```python
# img_b.shape
```

```python
img_recover = cv2.merge((img_b, img_g, img_r))
```

```python
plt.imshow(img_recover)
```

```python
img_border = cv2.copyMakeBorder(img, 10, 10, 10, 10, cv2.BORDER_CONSTANT, value=(255, 0, 0))
```

```python
plt.imshow(img_border)
```

```python
cv2.setUseOptimized(False)
e1 = cv2.getTickCount()
for k in range(5, 49, 2):
    cv2.medianBlur(img, k)
e2 = cv2.getTickCount()
t = (e2-e1)/cv2.getTickFrequency()
print('not optimized cost time:', t)

cv2.setUseOptimized(True)
e1 = cv2.getTickCount()
for k in range(5, 49, 2):
    cv2.medianBlur(img, k)
e2 = cv2.getTickCount()
t = (e2-e1)/cv2.getTickFrequency()
print('optimized cost time:', t)
```
