```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
import time
```

```python
def bgr_to_rgb(bgr):
    rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
    return rgb
def rgb_to_bgr(rgb):
    bgr = cv2.cvtColor(rgb, cv2.COLOR_RGB2BGR)
    return bgr
```

```python
# face_cascade = cv2.CascadeClassifier('../data/lbpcascades/lbpcascade_frontalface.xml')
face_cascade = cv2.CascadeClassifier('../data/lbpcascades/lbpcascade_frontalface_improved.xml')
```

```python
img = cv2.imread('../data/two_people.jpg')
```

```python
start_time = time.time()
faces = face_cascade.detectMultiScale(img)
end_time = time.time()
cost_time = end_time - start_time
print('cost:', cost_time)
```

```python
for face in faces:
    x, y, w, h = face
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
    
plt.imshow(bgr_to_rgb(img))
plt.axis('off')
```
